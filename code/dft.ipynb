```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate a random signal
np.random.seed(0)  # Seed for reproducibility

# Number of samples
random_signal = np.random.randn(1024)
N = len(random_signal)

assert len(random_signal) == N
```

```python
# Compute the DFT
X = np.zeros(N, dtype=complex)
for k in range(N):
    for n in range(N):
        X[k] += random_signal[n] * np.exp(-2j * np.pi * k * n / N)
```

```python

# Plot the random signal
plt.subplot(1, 2, 1)
plt.plot(random_signal)
plt.title('Random Signal in Time Domain')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
```
