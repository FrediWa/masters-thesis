```python
N = len(signal)
k = 5
PI = 3.141592653
Xk = 0
for n in range(N):
        Xk += signal[n] * np.exp(-2j * PI * k * n / N)
Xk
```

```python
# Compute the DFT
N = len(signal)
X = np.zeros(N, dtype=complex)

for k in range(N): 
    for n in range(N): 
        X[k] += signal[n] * cmath.exp(-2j * PI * k * n / N) 
        

X[25]
```

```python
from scipy.fft import fft, fftfreq
t = np.arange(0, N/100, 1/100)  # Time vector

# Step 1: Perform FFT
fft_values = fft(signal)

# Step 2: Compute the frequency bins
frequencies = fftfreq(N, 1/100)

# Step 3: Calculate the magnitude spectrum
magnitude = np.abs(fft_values)

# Plotting
plt.figure(figsize=(20, 5))

# Subplot 1: Original Signal
plt.subplot(1, 2, 1)
plt.plot(t, signal, color='g')
plt.title('Time-domain')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.grid()
plt.legend()

# Subplot 2: Magnitude Spectrum
plt.subplot(1, 2, 2)
plt.plot(frequencies[:N//2], magnitude[:N//2], color='b')
plt.title('Frequency-domain')
plt.xlabel('frequency')
plt.ylabel('magnitude')
plt.grid()
plt.legend()
```
